Key insight: PyTorch's Autograd automatically computes gradients, saving the need for manual calculus. The zero gradients for w and x show that when inputs are too large, sigmoid "maxes out" and loses the ability to learn — a good reminder that initial values matter.

```python
# Basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
```

```python
# Variables
w = torch.tensor(2.0, requires_grad=True)
x = torch.tensor(4.0, requires_grad=True)
b = torch.tensor(1.5, requires_grad=True)
```

```python
# main formula
z = torch.sigmoid(w * (x**2)) + (1 / (b**3))
print(f"z = {z.item()}")
```

```python
z.backward()
```

```python

print ("w.grad =", w.grad.item())
print ("x.grad =", x.grad.item())
print ("b.grad =", b.grad.item())
```
