```python
from datetime import date
```

```python
# Loan parameters
principal = 100_000
annual_rate = 0.12
start_date = date(2024, 1, 1)
```

```python
# Payment log: (payment_date, amount)
payment_log = [
    (date(2024, 1, 31), 5_000),
    (date(2024, 2, 29), 5_000),  # Leap Day — exists in 2024
    (date(2024, 3, 31), 5_000),
]

print("=" * 70)
print("CALENDAR-CORRECT AMORTIZATION AUDIT")
print(f"Principal: ${principal:,.2f} | Annual Rate: {annual_rate*100:.1f}%")
print(f"Start Date: {start_date}  (2024 is a leap year)")
print("=" * 70)

balance = principal
prev_date = start_date
total_interest = 0

for payment_date, payment in payment_log:
    days_elapsed = (payment_date - prev_date).days  # Exact calendar days
    interest = balance * (annual_rate / 365) * days_elapsed
    closing_balance = balance + interest - payment
    total_interest += interest

    print(f"\nPeriod : {prev_date} → {payment_date}")
    print(f"  Days elapsed     : {days_elapsed}  (exact calendar days)")
    print(f"  Opening balance  : ${balance:>12,.2f}")
    print(f"  Interest accrued : ${interest:>12,.4f}  "
          f"= {balance:,.2f} × (0.12/365) × {days_elapsed}")
    print(f"  Payment          : ${payment:>12,.2f}")
    print(f"  Closing balance  : ${closing_balance:>12,.2f}")

    prev_date = payment_date
    balance = closing_balance

print("\n" + "=" * 70)
print(f"Total interest accrued : ${total_interest:,.4f}")
print(f"Ending balance         : ${balance:,.2f}")
print("=" * 70)
```

```python
# --- Show the LLM trap ---
print("\n\nLLM TRAP COMPARISON (assumes 30 days/month)")
print("-" * 70)

llm_balance = principal
llm_periods = [
    ("Jan 1 → Jan 31", 30, 5_000),
    ("Jan 31 → Feb 29", 30, 5_000),  # LLM misses leap year — uses 30, not 29
    ("Feb 29 → Mar 31", 30, 5_000),  # LLM misses March's 31 days — uses 30
]
llm_total_interest = 0

for label, days, payment in llm_periods:
    interest = llm_balance * (annual_rate / 365) * days
    llm_total_interest += interest
    llm_balance = llm_balance + interest - payment
    print(f"  {label}: {days} days | Interest = ${interest:.4f}")

print(f"\nLLM ending balance         : ${llm_balance:,.2f}")
print(f"Calendar-correct balance   : ${balance:,.2f}")
print(f"Discrepancy (overcharge)   : ${llm_balance - balance:,.4f}")
print("\nVerdict: The LLM overcharges because it assumes Feb has 30 days")
print("(not 29) in a leap year, accruing extra interest for 1 phantom day.")
```
