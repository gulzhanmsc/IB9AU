```python
# Step 2: Load the model
from smolagents import CodeAgent, TransformersModel
import torch

print("⬇️ Loading Qwen 2.5 Coder 3B...")

model = TransformersModel(
    model_id="Qwen/Qwen2.5-Coder-3B-Instruct",
    device_map="auto",
    torch_dtype=torch.float16,
    max_new_tokens=2048
)

print("✅ Model loaded!")
```

```python
# Step 3: Instantiate the CodeAgent
agent = CodeAgent(tools=[], model=model, name="QuantitativeAnalyst")

# Step 4: Run the task prompt
task_prompt = """
Use yfinance, pandas, numpy, and matplotlib.pyplot to do the following:

1. Download daily closing prices for the last 180 days for these stocks:
   - Tech: ['NVDA', 'AAPL', 'MSFT']
   - Banks: ['JPM', 'BAC', 'C']
   Use: yf.download(['NVDA','AAPL','MSFT','JPM','BAC','C'], period='180d')['Close']

2. Calculate daily returns for each stock:
   returns = prices.pct_change().dropna()

3. Calculate the Sharpe Ratio for each stock (risk-free rate = 0):
   sharpe = (returns.mean() / returns.std()) * (252 ** 0.5)

4. Plot a bar chart of the 6 Sharpe Ratios:
   - Bars for NVDA, AAPL, MSFT must be colored GREEN
   - Bars for JPM, BAC, C must be colored BLUE
   - Title: 'Sharpe Ratio: Big Tech vs Big Banks (180 Days)'
   - X-axis label: 'Company'
   - Y-axis label: 'Annualized Sharpe Ratio'
   - Add the numeric Sharpe value on top of each bar, rounded to 2 decimal places
   - Add a legend: green patch = 'Big Tech', blue patch = 'Big Banks'

5. Save the chart as 'sharpe_comparison.png' using plt.savefig('sharpe_comparison.png', bbox_inches='tight')
   Then call plt.show()

6. Print each stock's Sharpe Ratio and state which sector has the higher average Sharpe Ratio.
"""

print("🤖 Agent is working...")
result = agent.run(task_prompt, stream=False)
print(result)

# Step 5: Display the saved chart
import IPython.display
import os

if os.path.exists("sharpe_comparison.png"):
    print("\n📊 Sharpe Ratio Comparison Chart:")
    IPython.display.display(IPython.display.Image("sharpe_comparison.png"))
else:
    print("⚠️ Chart not saved. Check agent output above for errors.")
```
